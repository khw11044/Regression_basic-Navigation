### 12. 위 의사결정나무(decision tree) 모델의 성능을 평가하려고 합니다.
### 예측 결과의 mae(Mean Absolute Error)를 구하세요.
* 성능 평가는 검증 데이터셋을 활용하세요.
* 11번 문제에서 만든 의사결정나무(decision tree) 모델로 y값을 예측(predict)하여 y_pred에 저장하세요.
* 검증 정답(y_valid)과 예측값(y_pred)의 mae(Mean Absolute Error)를 구하고 dt_mae 변수에 저장하세요.

```python
# 여기에 답안코드를 작성하세요.
from sklearn.metrics import mean_absolute_error

y_pred = model.predict(X_valid)
dt_mae = mean_absolute_error(y_valid, y_pred)
print('MAE:', dt_mae)
```

```python
def eval_graph(y_valid, y_pred):
    plt.figure(figsize=(18,6))
    plt.plot(y_valid.values[:250], label='Actual',  marker = 'o')
    plt.plot(y_pred[:250], label='Predicted', alpha = 0.8,  marker = 'o')
    plt.legend()
    plt.ylabel('Time_Driving(실주행시간)')
    plt.show()

eval_graph(y_valid, y_pred)
```

```python
## H Percentage of Correct ETA : HPCETA 평가지표 
import numpy as np
def HPCETA(y_valid, y_pred, threshold=120):

    error = y_valid.values - y_pred.reshape(-1)
    per = np.where(error<=threshold, 1, 0)
    
    return sum(per) / len(per)
    
print(HPCETA(y_valid, y_pred, threshold=120))
```

### 13. Time_Driving(실주행시간)을 예측하는 딥러닝 모델을 만들려고 합니다.
### 아래 가이드에 따라 모델링하고 학습을 진행하세요.
* Tensoflow framework를 사용하여 딥러닝 모델을 만드세요.
* 히든레이어(hidden layer) 2개이상으로 모델을 구성하세요.
* dropout 비율 0.2로 Dropout 레이어 1개를 추가해 주세요.
* 손실함수는 MSE(Mean Squared Error)를 사용하세요.
* 하이퍼파라미터 epochs: 30, batch_size: 16으로 설정해주세요.
* 각 에포크마다 loss와 metrics 평가하기 위한 데이터로 X_valid, y_valid 사용하세요.
* 학습정보는 history 변수에 저장해주세요

```python
X_train.shape
```

```python
# 여기에 답안코드를 작성하세요.

model = Sequential()

model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(16, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse',metrics=['mae','mse'])
# es = EarlyStopping(monitor='val_loss', patience=4, mode='min', verbose=1)    # val_loss
mc = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True, verbose=1)

history = model.fit(X_train, y_train, 
                    batch_size=16, 
                    epochs=30, 
                    callbacks=[mc],
                    validation_data=(X_valid, y_valid), 
                    verbose=1).history
```

```python
# 함수로 만들어서 사용합시다.
def dl_history_plot(history):
    plt.figure(figsize=(16,4))
    plt.subplot(1,3,1)
    plt.plot(history['loss'], label='train_err', marker = '.')
    plt.plot(history['val_loss'], label='val_err', marker = '.')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid()
    
    plt.subplot(1,3,2)
    plt.plot(history['mse'], label='Train', marker = '.')
    plt.plot(history['val_mse'], label='Validation', marker = '.')
    plt.ylabel('MSE')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid()
    
    plt.subplot(1,3,3)
    plt.plot(history['mae'], label='Train', marker = '.')
    plt.plot(history['val_mae'], label='Validation', marker = '.')
    plt.ylabel('MAE')
    plt.xlabel('Epoch')
    plt.legend()
    plt.grid()
    
    plt.show()
```

```python
# 여기에 답안코드를 작성하세요.
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler

target = 'Time_Driving'

x = df_preset.drop(target, axis=1)
y = df_preset.loc[:,target]

X_train, X_valid, y_train, y_valid = train_test_split(x,y, test_size=0.2, random_state=42)
print(X_train.shape, X_valid.shape, y_train.shape, y_valid.shape)


# 정규화
col_name = list(X_train)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
```

```python
model = Sequential()

model.add(Dense(128, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(16, activation='relu'))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse',metrics=['mae','mse'])
# es = EarlyStopping(monitor='val_loss', patience=4, mode='min', verbose=1)    # val_loss
mc = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True, verbose=1)

history = model.fit(X_train, y_train, 
                    batch_size=16, 
                    epochs=30, 
                    callbacks=[mc],
                    validation_data=(X_valid, y_valid), 
                    verbose=1).history
```
